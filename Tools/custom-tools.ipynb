## Custom Tool

### Using `tool` Decorator

```python
from langchain_core.tools import tool
```

```python
# Step 1 - Create a function
def multiply(a, b):
    """Multiply two numbers"""
    return a * b
```

```python
# Step 2 - add type hints

def multiply(a: int, b: int):
    """Multiply two numbers"""
    return a * b
```

```python
# Step 3 - add tool decorator

@tool
def multiply(a: int, b: int):
    """Multiply two numbers"""
    return a * b
```

```python
result = multiply.invoke({"a": 3, "b": 2})
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema())
```

### Using `StructuredTool` and `Pydantic`

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(..., description="First number to multiply")
    b: int = Field(..., description="Second number to multiply")
```

```python
def multiply(a: int, b: int) -> int:
    return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply,
    name="multiply",
    description="Multiply two numbers",
    args_schema=MultiplyInput,
)
```

```python
result = multiply_tool.invoke({"a": 3, "b": 2})
print(result)
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

```python
print(multiply_tool.args_schema.model_json_schema())
```

### Using `BaseTool`

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
# Args schema using Pydantic
class MultiplyInput(BaseModel):
    a: int = Field(..., description="First number to multiply")
    b: int = Field(..., description="Second number to multiply")
```

```python
class MultiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers"

    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({"a": 3, "b": 2})
print(result)
```
