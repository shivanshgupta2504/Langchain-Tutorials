```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
# tool
@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI()
```

```python
llm_with_tools = llm.bind_tools(tools=[multiply]) # tool binding
```

```python
llm_with_tools.invoke("Hi are you?")
```

```python
query = HumanMessage(content="Can you multiply 3 with 15")
```

```python
messages = [query]
```

```python
messages
```

```python
response = llm_with_tools.invoke(messages)
response
```

```python
messages.append(response)
```

```python
response.tool_calls[0]
```

```python
# Tool Execution
tool_call = response.tool_calls[0]
tool_result = multiply.invoke(tool_call) # Tool execution
```

```python
tool_result
```

```python
messages.append(tool_result)
```

```python
messages
```

```python
llm_with_tools.invoke(messages)
```
